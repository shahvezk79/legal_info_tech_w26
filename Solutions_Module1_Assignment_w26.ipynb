### Beginner Question 2:

In 1875, the Supreme Court of Canada was established under the Supreme Court Act. In 1982, Bertha Wilson became the first woman appointed to the Supreme Court of Canada. Assume for the sake of simplicity that both events happened on the same day of the year.
In the code cell below:

Create a variable storing the year the Supreme Court of Canada was established.
Create a variable storing the year the first woman was appointed to the Supreme Court.
Create a variable that calculates the number of years between these two events.
Create a variable that calculates the number of months between these two events.
Print: "It took # months from the establishment of the Supreme Court of Canada for a woman to be appointed to the bench."

```python
year_established = 1875
year_appointed = 1982
years_between = year_appointed - year_established
months_between = years_between * 12

# Various ways to print
print(f"It took {months_between} months from the establishment of the Supreme Court of Canada for a woman to be appointed to the bench.")
print("It took", months_between, "months from the establishment of the Supreme Court of Canada for a woman to be appointed to the bench.")
print("It took " + str(months_between) + " months from the establishment of the Supreme Court of Canada for a woman to be appointed to the bench.")

```

```python
# Alternative approach that can be more easily adapted to complex scenarios:
# Credit to Amelia Cox (with revisions)

from datetime import datetime

start_date = datetime(1875, 1, 1)
end_date = datetime(1982, 1, 1)

elapsed_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)

print(f"It took {elapsed_months} months from the establishment of the Supreme Court of Canada for a woman to be appointed to the bench.")
```

### Beginner Question 3:

The code cell below includes a list of current Supreme Court of Canada judges. Using this list, produce the exact output below using the print function. Note that the Chief Justice has a different honorific. Desired Output:

The Right Honourable Chief Justice Richard Wagner

The Honourable Justice Andromache Karakatsanis

The Honourable Justice Suzanne Côté

The Honourable Justice Malcolm Rowe

The Honourable Justice Sheilah Martin

The Honourable Justice Nicholas Kasirer

The Honourable Justice Mahmud Jamal

The Honourable Justice Michelle O'Bonsawin

The Honourable Justice Mary Moreau

```python
list_of_judges = [
    "Richard Wagner",
    "Andromache Karakatsanis",
    "Suzanne Côté",
    "Malcolm Rowe",
    "Sheilah Martin",
    "Nicholas Kasirer",
    "Mahmud Jamal",
    "Michelle O'Bonsawin",
    "Mary Moreau"
]
```

```python
for judge in list_of_judges:
  if judge == "Richard Wagner":
    print("The Right Honourable Chief Justice " + judge)
  else:
    print("The Honourable " + judge)
```

```python
# Alternative approach based on order of list
# Credit to Chong Tan (with revisions)

print(f"The Right Honourable Chief Justice {list_of_judges[0]}")
for judge in list_of_judges[1:]:
    print(f"The Honourable Justice {judge}")
```

### Intermediate Question 2:

The code cell below includes a dictionary with the highest neutral citation for Federal Court cases for several years. For example in 2020, neutral citations go from "2020 FC 1" to "2020 FC 6805". In other words, the first federal court case in 2020 is "2020 FC 1", the second is "2020 FC 2", and so on until "2020 FC 6805".

Create a variable with a dictionary that uses each year from 2013 to 2022 as keys, and the values associated with those keys should be a list of all the neutral citations for that year. You can imagine that this dictionary would be used to scrape each decision from the federal court website.


```python
highest_neutral_citation =  {
    2013: 8428,
    2014: 8508,
    2015: 5901,
    2016: 5412,
    2017: 5641,
    2018: 6616,
    2019: 7910,
    2020: 6805,
    2021: 9832,
    2022: 13672
}

citation_dictionary = {}
for year, citation in highest_neutral_citation.items():
  year_list = []
  for citation_number in range(1, citation + 1):
    year_list.append(f"{year} FC {citation_number}")
  citation_dictionary[year] = year_list
```

```python
# Or a more advanced pythonic approach using dictionary comprehension:

citation_dictionary = {
    year: [f"{year} FC {i}" for i in range(1, citation + 1)]
    for year, citation in highest_neutral_citation.items()
}
```
