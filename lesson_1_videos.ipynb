### Conditions

```python
# ==, !=, <, >, <=, >=, in, not in

A = 10
B = 20
C = 15

if A == B:
    print("A is equal to B")
    print ("Great!")
elif A == C:
    print("A is equal to C")
else:
    print("A is not equal to B or C")

A = "abc"
B = "abc"

if A == B:
    print("A is equal to B")
    print ("Great!")
else:
    print("A is not equal to B")

A = "def"

if "a" not in A:
    print("A does not contain the letter a")
else:
    print("A does contain the letter a")

year_is_2023 = False

if year_is_2023:
    print("It is 2023")
else:
    print("It is not 2023")



```
