### Beginner Question 2:

Using the [Canadian Human Rights Tribunal dataset](https://huggingface.co/datasets/a2aj/canadian-case-law) from the A2AJ, make a line chart with the average number Canadian Human Rights Tribunal decisions per year over the past 10 years.

```python
# pip install pandas pyarrow matplotlib
# (Optional) If you want to use Hugging Face Datasets too: pip install datasets

import pandas as pd
import matplotlib.pyplot as plt

# 1) Load the parquet into a DataFrame
parquet_url = "https://huggingface.co/datasets/a2aj/canadian-case-law/resolve/main/CHRT/train.parquet"
df = pd.read_parquet(parquet_url)

# 2) Parse document_date_en safely (timezone-aware)
df["document_date_en"] = pd.to_datetime(df["document_date_en"], utc=True, errors="coerce")

# Drop rows with missing/invalid dates
df = df.dropna(subset=["document_date_en"])

# 3) Filter to the past 10 years (based on max date in dataset)
end_dt = df["document_date_en"].max()
start_dt = end_dt - pd.DateOffset(years=10)

df_10y = df.loc[(df["document_date_en"] >= start_dt) & (df["document_date_en"] <= end_dt)].copy()

# 4) Count decisions per year
df_10y["year"] = df_10y["document_date_en"].dt.year

decisions_per_year = (
    df_10y.groupby("year")
    .size()
    .sort_index()
)

# Optional: ensure all years in range appear (even if 0 decisions)
all_years = pd.Index(range(start_dt.year, end_dt.year + 1), name="year")
decisions_per_year = decisions_per_year.reindex(all_years, fill_value=0)

# 5) Line chart (yearly counts) + optional horizontal average
avg_per_year = decisions_per_year.mean()

plt.figure()
plt.plot(decisions_per_year.index, decisions_per_year.values, marker="o")
plt.axhline(avg_per_year, linestyle="--")  # average across the 10-year window

plt.title("CHRT decisions per year (past 10 years)")
plt.xlabel("Year")
plt.ylabel("Number of decisions")
plt.xticks(decisions_per_year.index, rotation=45)
plt.tight_layout()
plt.show()

print("Average decisions/year over this window:", avg_per_year)
```

### Beginner Question 3:

Using the A2AJ's [Canadian Case Law](https://huggingface.co/datasets/a2aj/canadian-case-law) dataset, pick two courts/tribunals, and create line charts for the number of words per decision per year for the past five years for each of the two courts/tribunals. What is the most interesting change over this time that you observe?


```python
import pandas as pd
import matplotlib.pyplot as plt

CHRT_URL = "https://huggingface.co/datasets/a2aj/canadian-case-law/resolve/main/CHRT/train.parquet"
BCCA_URL = "https://huggingface.co/datasets/a2aj/canadian-case-law/resolve/main/BCCA/train.parquet"

def yearly_avg_words_english_only(parquet_url: str, label: str) -> pd.DataFrame:
    df = pd.read_parquet(parquet_url)

    date_col = "document_date_en"
    text_col = "unofficial_text_en"

    if date_col not in df.columns:
        raise KeyError(f"{label}: missing {date_col}. Columns: {list(df.columns)}")
    if text_col not in df.columns:
        raise KeyError(f"{label}: missing {text_col}. Columns: {list(df.columns)}")

    df[date_col] = pd.to_datetime(df[date_col], utc=True, errors="coerce")
    df = df.dropna(subset=[date_col])

    # English-only: drop empty/blank English text
    text = df[text_col].fillna("").astype(str).str.strip()
    df = df.loc[text.ne("")].copy()
    text = text.loc[df.index]

    df["year"] = df[date_col].dt.year
    df["word_count"] = text.str.split().str.len()

    out = (
        df.groupby("year")["word_count"]
        .agg(avg_words="mean", n_decisions="size")
        .reset_index()
    )
    out["tribunal"] = label
    return out

chrt = yearly_avg_words_english_only(CHRT_URL, "CHRT")
bcca = yearly_avg_words_english_only(BCCA_URL, "BCCA")

# Define window: last 5 years up to latest year present across both (English dates only)
max_year = int(max(chrt["year"].max(), bcca["year"].max()))
years = list(range(max_year - 4, max_year + 1))

chrt5 = chrt.loc[chrt["year"].isin(years)].set_index("year")
bcca5 = bcca.loc[bcca["year"].isin(years)].set_index("year")

# Reindex so missing years show as gaps
chrt5 = chrt5.reindex(years)
bcca5 = bcca5.reindex(years)

# Optional: ignore years with too-few decisions (reduces noisy spikes)
MIN_N = 5
chrt_plot = chrt5["avg_words"].where(chrt5["n_decisions"] >= MIN_N)
bcca_plot = bcca5["avg_words"].where(bcca5["n_decisions"] >= MIN_N)

# Plot
plt.figure()
plt.plot(years, chrt_plot.values, marker="o", label="CHRT")
plt.plot(years, bcca_plot.values, marker="o", label="BCCA")

plt.title(f"Average words per decision per year (English-only), last 5 years up to {max_year}")
plt.xlabel("Year")
plt.ylabel("Avg words per decision")
plt.xticks(years, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Quick sanity check printouts
print("Years plotted:", years)
print("\nCHRT decisions/year (English text, non-empty):")
print(chrt5["n_decisions"])
print("\nBCCA decisions/year (English text, non-empty):")
print(bcca5["n_decisions"])
```

The most interesting trend I noticed here is how consistent the BCCA has stayed in their word count over the years, growing slightly more concise but remaining stable.

### Intermediate Question 1:

Using the [Tax Court of Canada dataset](https://huggingface.co/datasets/a2aj/canadian-case-law) from A2AJ, print the names of the five Tax Court of Canada cases that have most frequently been cited by the Tax Court of Canada over the past 10 years.

Hint: neutral citations follow specific patterns (e.g. YYYY TCC #).

```python
import pandas as pd

# 1) Load only the columns we need (faster, less memory)
TCC_URL = "https://huggingface.co/datasets/a2aj/canadian-case-law/resolve/main/TCC/train.parquet"

cols = [
    "name_en",
    "citation_en",
    "citation2_en",
    "document_date_en",
    "unofficial_text_en",
]
df = pd.read_parquet(TCC_URL, columns=cols)

# 2) Parse dates, filter to past 10 years (relative to dataset max date)
df["document_date_en"] = pd.to_datetime(df["document_date_en"], utc=True, errors="coerce")
df = df.dropna(subset=["document_date_en"])

end_dt = df["document_date_en"].max()
start_dt = end_dt - pd.DateOffset(years=10)

df_10y = df.loc[(df["document_date_en"] >= start_dt) & (df["document_date_en"] <= end_dt)].copy()

# 3) Build citation -> name mapping (from the dataset itself)
map_parts = []
if "citation_en" in df.columns:
    map_parts.append(df[["citation_en", "name_en"]].dropna(subset=["citation_en"]).rename(columns={"citation_en": "citation"}))
if "citation2_en" in df.columns:
    map_parts.append(df[["citation2_en", "name_en"]].dropna(subset=["citation2_en"]).rename(columns={"citation2_en": "citation"}))

citation_to_name = (
    pd.concat(map_parts, ignore_index=True)
      .dropna(subset=["citation"])
      .drop_duplicates(subset=["citation"], keep="first")
      .set_index("citation")["name_en"]
)

# 4) Extract neutral citations from English text and count *decisions citing* each case
# Neutral citation pattern example: "2021 TCC 37"
pattern = r"\b(19|20)\d{2}\s+TCC\s+\d+\b"

# Find citations per decision, de-duplicate within each decision
found = (
    df_10y["unofficial_text_en"]
      .fillna("")
      .astype(str)
      .str.findall(pattern)   # NOTE: because pattern has a capturing group, this returns only the group!
)

# Fix: use a non-capturing group so findall returns full matches
pattern_full = r"\b(?:19|20)\d{2}\s+TCC\s+\d+\b"
found = (
    df_10y["unofficial_text_en"]
      .fillna("")
      .astype(str)
      .str.findall(pattern_full)
      .apply(lambda xs: sorted(set(xs)))  # unique per decision
)

# Explode and count how many decisions cite each neutral citation
cited_counts = (
    found.explode()
         .dropna()
         .value_counts()
         .head(5)
         .rename_axis("citation")
         .reset_index(name="n_decisions_citing")
)

# 5) Attach names (if available) and print
cited_counts["case_name_en"] = cited_counts["citation"].map(citation_to_name)

print(f"Window (dataset-relative): {start_dt.date()} to {end_dt.date()}\n")
print(cited_counts.to_string(index=False))
```

Load into a dataframe the excel data that reports tabular data about Supreme Court of Canada decisions discussed in the Module 2 videos available at this URL:

    https://litigate.com/assets/uploads/20230823-140809-6601-SCC-Database-v-2023.01.xlsx

For the meaning of the values of codes in the excel data, see the coding manual [here](https://litigate.com/assets/uploads/20220216-111705-4622-Lenczner-Slaght-Supreme-Court-of-Canada-Database-Coding-Manual-v-2022.01.pdf).

Print a list the names of the 5 judges who wrote the largest number of dissenting opinions in the dataset, and then print a bar chart with their names and the number of dissenting opinions in the dataset that each of them wrote.

Do you have any theories as to why these 5 judges in particular wrote the largest number of dissenting opinions?

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset
# We use the URL provided. The relevant data is in the 'Case' sheet.
url = "https://litigate.com/assets/uploads/20230823-140809-6601-SCC-Database-v-2023.01.xlsx"
df_case = pd.read_excel(url, sheet_name='Case')
df_justices = pd.read_excel(url, sheet_name='JusticesPresent')

# 2. Map Justice IDs to Names
# We create a dictionary from the JusticesPresent sheet to decode the IDs.
# 'justiceID' corresponds to the codes in the PDF and the dissentWriter column.
justice_mapping = df_justices[['justiceID', 'justiceName']].drop_duplicates().set_index('justiceID')['justiceName'].to_dict()

# 3. Count Dissenting Opinions
# We count the occurrences of each ID in the 'dissentWriter' column.
# We exclude '0', which indicates no dissent.
dissent_counts = df_case[df_case['dissentWriter'] != 0]['dissentWriter'].value_counts()

# 4. Get the Top 5
top_5_dissenters = dissent_counts.head(5)

# Prepare lists for plotting
names = [justice_mapping.get(jid, f"ID {jid}") for jid in top_5_dissenters.index]
counts = top_5_dissenters.values

print("Top 5 Judges by Number of Dissenting Opinions:")
for name, count in zip(names, counts):
    print(f"{name}: {count}")

# 5. Create Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(names, counts, color='#4c72b0', zorder=3)

# Formatting
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
plt.ylabel("Number of Dissents")
plt.title("Top 5 Supreme Court Justices by Dissenting Opinions")
plt.xticks(rotation=45)

# Add counts on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
